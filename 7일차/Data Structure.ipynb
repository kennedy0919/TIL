```python
# 혈액형 인원수 세기
# 결과 => {"A" : 3, "B" : 3, "O" : 3, "AB" : 3}
blood_types = ["A", "B", "A", "O", "AB", "AB","A","O", "B", "O", "B", "AB"]

#[]
new_dict = {}
# bllod_type을 순회하면서
for blood_type in blood_types:  
    # 기존에 키가 이미 존재한다면, 
    if blood_type in new_dict:
        # 기존에 키의 값을 +1 증가
        new_dict[blood_type] += 1
    # 키가 존재하지 않는다면(처음 설정되는 키)
    else:
        new_dict[blood_type] = 1
print(new_dict)

# .get
new_dict = {}
# bllod_type을 순회하면서
for blood_type in blood_types:
    new_dict[blood_type] = new_dict.get(blood_type, 0) + 1

print(new_dict)

# .setdefault()
new_dict = {}
for blood_type in blood_types:
    new_dict.setdefault(blood_type, 0)
    new_dict[blood_type] += 1
print(new_dict)
```

```python
a = [1, 2, 3]

# 슬라이싱
b = a[:]
b[0] #[1, 2, 3][100, 2, 3]

# copy
c = a.copy()
c[0] = 100
print(a, c) #[1, 2, 3][100, 2, 3]

##### 얕은 복사의 한계
a = [1, 2, [1, 2]]

b = a[:]
b[2][0] = 999
print(a, b)

c = a.copy()
c[2][0] = 999
print(a, c)


# 깊은 복사
import copy

original_list = [1, 2, [1, 2]]
deep_copied_list = copy.deepcopy(original_list)

deep_copied_list[2][0] = 999

print(original_list, deep_copied_list)
```
