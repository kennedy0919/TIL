# 리스트 값 추가 및 삭제 메서드
1. .append 
    - 리스트 마지막에 항복 추가

2. extend(iterable)
    - 리스트에 다른 반복 가능한 객체의 모든 항목을 추가

```python
## append 와 extend 차이

numbers = [1, 2, 3]
numbers2 = [4, 5, 6]
numbers.append(numbers2)
print(numbers)

numbers.extend(numbers)
print(numbers)
```

### .pop
- 리스트에서 지정한 인덱스의 항목을 제거하고 반환 
- 작성하지 않을 경우 마지막 항목을 제거

```python
numbers = [1, 2, 3]
# sort 메서드
print(numbers.sort())   # None

numbers = [3, 2, 1]
# sorted 함수
print(sorted(numbers))  # [1, 2, 3]
print(numbers)  # [3, 2, 1]
```

```python
#### quize

numbers = [1, 2, 3]

# 1. 할당
list1 = numbers

# 2. 슬라이싱
list2 = numbers[:]

numbers[0] = 100
print(list1)  ## [100, 2, 3]
print(list2)  ## [1, 2, 3]
```
