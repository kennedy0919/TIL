# 조건문
- 주어진 조건식을 평가하여 해당 조건이 참인 경우에만 코드 블록을 실행하거나 건너뜀

if/ elif / else
if, elif 가 모두 거짓이라면 else 실행
else 는 선택적으로 실행가능


```python
num = int(input("숫자를 입력하세요 :"))

# if statement
# num이 홀수라면(2로 나눈 나머지가 1이라면)
# if num % 2:
# 이것도 가능한 이유는 if 1: 은 True 로 평가함
if num % 2 == 1:
    print("홀수입니다.")

# num이 홀수가 아니라면(짝수면)
else:
    print("짝수입니다.")
```

# 반복문
- 주어진 코드 블록을 여러번 반복해서 실행하는 구문
- 특정 작업을 반복적으로 수행, 주어진 조건이 참인 동안 반복적으로 수행(False가 되면 멈춤)

# for
- 임의의 시퀀스의 항목들을 그 시퀀스에 들어있는 순서대로 반복
- for 변수 in 반복 가능한 객체:

- 반복 가능한 객체 : 시퀀스, dict, set

# for 문 원리
- 리스트 내 첫 항목이 반복 변수에 할당되고 코드 블록이 실행
- 다음으로 반복 변수에 리스트의 2번째 항목이 할당되고 코드블록이 다시 실행
- ...마지막으로 반복 변수에 리스트의 마지막 요소가 할당되고 코드블록이 실행

# 적절한 반복문 활용하기
1. for
    - 반복 횟수가 명확하게 정해져 있는 경우에 유용
    - 예를 들어 리스트, 튜플, 문자열 등과 같은 시퀀스 형식의 데이터를 처리할 때

2. while

# 반복 제어
- for문과 while 은 매 반복마다 본문 내 모든 코드를 실행하지만 때때로 일부만 실행하는 것이 필요할 때가 있음

1. break : 반복을 즉시 중지
2. continue : 다음 반복으로 건너뜀

# List Comprehension
- [표현식 for 변수 in 반복 가능한 객체]
- list(표현식 for 변수 in 반복 가능한 객체) 

```python
## 0~9 요소를 가지는 리스트 만들기(1~10)
new_list = []
for i in range(10):
    new_list.append(i)
print(new_list)

## List Comprehension
new_list_2 = [i for i in range(10)]
print(new_list_2)

## 0~9 에서 홀수만 가지는 리스트 만들기(1~10)
new_list = []
for i in range(10):
    if i % 2 == 1:
        new_list.append(i)
print(new_list)

## List Comprehension
new_list_2 = [i for i in range(10) if i % 2 == 1]
print(new_list_2)
```

```python
## 반복문에서 else를 사용하는 경우
## for - else
for i in range(40):
    pass
else:
    pass

# else 문은 반복문을 실행한 이후에 실행됩니다.
# 반복문이 break 문으로 종료될때는 else 문이 실행되지 않습니다.
# break 를 통해 중간에 종료되지 않은 경우만 else 문이 실행된다.

# 반복문이 중간에 종료된 적이 있는지를 검사할때 유용하게 사용

string1 = "apple"
string2 = "banana"
for c in string1:
    # apple 문자열을 순회하면서 알파벳 b가 있으면 b를 출력한 다음 반복문을 종료
    if c == "b":
        print("b!")
        break
else :
    print("이 단어 안에는 b가 없습니다.")

for c in string2:
    # banana 문자열을 순회하면서 알파벳 b가 있으면 b를 출력한 다음 반복문을 종료
        if c == "b":
            print("b!")
        break
else :
    print("이 단어 안에는 b가 없습니다.")
```

```python
## 리스트를 생성하는 4가지 방법 비교
#3 정수 1,2,3, 을 가지는 새로운 리스트 만들기
## 어떤게 제일 빨라요?? -> 의미없음

numbers = ["1", "2", "3"]

# 1. for loop
new_numbers = []
for number in numbers:
    new_numbers.append(int(number))
print(new_numbers)

# 2. map
new_numbers_2 = list(map(int,numbers))
print(new_numbers_2)

# 3. list comprehension
new_numbers_3 = [int(number) for number in numbers]
print(new_numbers_3)
```

```python
# enumerate
result = ["a", "b", "c"]

print(enumerate(result)) 
print(list(enumerate(result)))

for index, elem in enumerate(result):
    print(index, elem)
```

```python
## pass 와 continue의 차이
for num in range(1, 11):
    if num % 2 == 1:
        continue
    print(num, end = " ")

## pass 와 continue의 차이
for num in range(1, 11):
    if num % 2 == 1:
        pass
    print(num, end = " ")
    # pass 는 아무역할을 하지 않음
    # continue는 명백한 기능이 있음
    # continue는 반복문을 건너뛰는 기능이 있음
```

```python
scores = {"minseok" : 100, "gildong" : 80, "jaedong" : 70}

# 기본적으로 key만 가져감
for name in scores:
    print(f"{name} : {scores[name]}")
```

```python
scores = {"minseok" : 100, "gildong" : 80, "jaedong" : 70}

# 기본적으로 key만 가져감
for score in scores.values():
    print(score)
    
```

```python
scores = {"minseok" : 100, "gildong" : 80, "jaedong" : 70}

# 기본적으로 key만 가져감
for key, value in scores.items():
    print(f"{key} : {value}")
```
