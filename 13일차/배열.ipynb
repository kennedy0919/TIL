# 2차원 배열의 선언
- 1차원 List를 묶어놓은 List
- 2차원 이상의 다차원 List는 차원에 따라 index를 선언
- 2차원 List의 선언 : 세로길이(행의개수), 가로길이(열의개수)가 필요함
- Python 에서는 데이터 초기화를 통해 변수선언과 초기화가 가능함

```python
# j가 증가하는 값에따라 감소하는 값 만들기

# 어떤 값에서 증가하는 값을 빼주면 됨

# arr = [[0] * M for _ in range(N)]  이거로 써야댐
# arr = [[0] * M] * N 으로 쓰면 안댐
```

```python
'''
3
1 2 3
4 5 6
7 8 9

'''
di = [0, 1, 0, -1]
dj = [1, 0, -1, 0]
N = int(input())
arr = [list(map(int, input().split())) for _ in range(N)]

max_v = 0  # 모든 원소가 0이상
for i in range(N):  # 모든 원소 arr[i][j]에 대해
    for j in range(N):
        # arr[i][j] 중심으로
        s = arr[i][j]
        for k in range(4):
            ni, nj = i+di[k], j+dj[k]
            if 0 < ni < N and 0 <= nj < N:
                s += arr[ni][nj]
        # 여기까지 주변 원소를 포함한 합
        if max_v < s: # s > max_v 이렇게 하지 말고 밑에 랑 좌우 맞춰주기
            max_v = s
```
