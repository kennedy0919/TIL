# Python 에서의 문자열 처리
- 문자열은 직접 수정이 불가능하다.


```python
s = "Reverse"
s_list = list(s)
print(s_list)

N = len(s)
print(N)

for i in range(0, N//2):
    s_list[i], s_list[N -1 -i] = s_list[N - 1 - i], s_list[i]

# print(s_list)
s = ''.join(s_list)
print(s)
```

```python
s1 = "abc"
s2 = "abc"
s5 = s1[0:2] + "c"
print(s1, s2, s5)

if s1 == s2:
    print("s1 == s5")
else:
    print("s1 != s5")
if s1 is s2:
    print("s1 is s5")
else:
    print("s1 is not s5")
```

```python
def itoa(a):
    s = ""
    while a > 0:
        s = s + chr(ord("0") + (a % 10)) # 1dmlwkfl tntwkdml ASCII 값
        a //= 10
    return s[::-1]

a = 123
print(itoa(a))
```
