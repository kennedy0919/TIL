# 큐
- 큐는 선입선출 구조

```python
# 원형큐
def enQ(data):
    global rear
    if (rear+1) % cQsize == front:
        print("cQ is Full")
    else:
        rear = (rear+1) % cQsize
        cQsize[rear] = data

def deQ():
    global front
    front = (front+1) % cQsize
    return cQ[front]


cQsize = 4
cQ = [0] * cQsize
front = 0
rear = 0

enQ(1)
enQ(2)
enQ(3)
enQ(4)
print(deQ())
print(deQ())
```
