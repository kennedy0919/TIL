```python
## 재귀함수 기본
def f(i, N):     # i 현재 상태, N 목표
    if i == N:
        return
    else:
        f(i+1, N)
```

```python
## 재귀함수 기본
def f(i, N):     # i 현재 상태, N 목표
    if i == N:
        print(B)
        return
    else:
        B[i] = A[i]
        f(i+1, N)
N = 5
A = [1,2,3,4,5]
B = [0] * N
f(0, N)
```

```python
## key가 있으면 1, 없으면 0을 리턴하는 함수
def f(i, N, key, arr):     # i 현재 상태, N 목표, key
    if i == N:
        return 0   # key가 없는 경우
    elif arr[i] == key:
        return 1

    else:
        B[i] = A[i]
        return f(i+1, N, key, arr)
N = 5
A = [1,2,3,4,5]
key = 10
B = [0] * N
print(f(0, N, key, A))
```

```python
## 6자리 숫자에 대해서 완전 검색을 이용하는법
def f(i, N):
    if i == N:    # 순열 완성
        return
    else:          # p[i]에 들어갈 숫자를 결정
        for j in range(0, N):
            if used[j] == 0:    # 아직 사용되기 전이면
                p[i] = card[j]
                used[j] = 1
                f(i+1, N)
                used[j] = 0

card = list(map(int, input()))
used = [0] * 6  # 이미 사용한 카드인지 표시
p = [0] * 6
f(0, 6)
```
