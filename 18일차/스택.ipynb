- 스택은 후입선출
- 가장 마지막에 저장된 위치 = stack pointer

# 괄호검사
- ​조건1 : 왼쪼 괄호 개수와 오른쪽 괄호의 개수가 같아야 한다.
- 조건2 : 같은 괄호에서 왼쪽 괄호는 오른쪽 괄호보다 먼저 나와야 한다.
- 조건3 : 괄호 사이에는 포함관계만 존재한다.

# 재귀함수
- 확정된값은 정해놓는다

```python
# 스택 사용 (파이썬의 메소드)
stack = []
def py_push(item):
    stack.append(item)
def py_pop():
    if len(stack) == 0:
        # 언더플로우!
        return
    else:
        return stack.pop()
for i in range(10):
    py_push(i)
print(stack)
for i in range(10):
    print(py_pop(), end=" ")
print()
print(stack)
print("==============================")
# 스택 사용 (인덱스)
top = -1  # 원소를 마지막으로 삽입한 위치
size = 10
stack = [0] * size
def my_push(item):
    global top
    top += 1
    if top == size:
        print("오버플로우")
    else:
        stack[top] = item
def my_pop():
    global top
    if top == -1:
        print("언더플로우")
        return
    else:
        top -= 1
        return stack[top + 1]
def peek():
    # top 이 -1 이면 원소가 없다
    if top > -1:
        return stack[top]
for i in range(10):
    my_push(i)
print(stack)
for i in range(10):
    print(my_pop(), end=" ")
print()
print(stack, top)
my_push(100)
print(stack, top)
```
